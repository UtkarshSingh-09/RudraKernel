```python
import json
from pathlib import Path

checkpoint_path = Path('artifacts/training/base_checkpoint.json')
metrics_path = Path('artifacts/training/base_metrics.json')

print('checkpoint exists:', checkpoint_path.exists())
print('metrics exists:', metrics_path.exists())

if checkpoint_path.exists():
    checkpoint = json.loads(checkpoint_path.read_text(encoding='utf-8'))
    print('non_zero_gradient_signal:', checkpoint.get('non_zero_gradient_signal'))

if metrics_path.exists():
    metrics = json.loads(metrics_path.read_text(encoding='utf-8'))
    print('reward_mean:', metrics.get('reward_mean'))
    print('reward_std:', metrics.get('reward_std'))
```
